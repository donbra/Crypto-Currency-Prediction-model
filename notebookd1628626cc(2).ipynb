```python
bitcoin_df_2['Profitable_day'] = bitcoin_df_2['Close'] > bitcoin_df_2['Open']
bitcoin_df_2['Return'] = (bitcoin_df_2['Close'].astype(float) - bitcoin_df_2['Open'].astype(float)) / bitcoin_df_2['Open'].astype(float)
bitcoin_df_2['Date'] = pd.to_datetime(bitcoin_df_2['Date'])
bitcoin_df_2['Year'] = bitcoin_df_2['Date'].dt.year
bitcoin_df_2['Close'] = bitcoin_df_2['Close'].astype(float)
bitcoin_df_2['Open'] = bitcoin_df_2['Open'].astype(float)
annual_returns = (bitcoin_df_2.groupby('Year')['Close'].agg(lambda x: x.iloc[-1] / x.iloc[0] - 1) * 100)
final_df = pd.DataFrame(annual_returns)
final_df = final_df.rename(columns={"Close": "Annual_Return"})
total_growth = pd.DataFrame({"Annual_Return":[final_df["Annual_Return"].sum()]}, index = ['Total'])
combined_df = pd.concat([final_df, total_growth])
print(combined_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
bitcoin_df_2.head()
bitcoin_df_2['Spread'] = bitcoin_df_2['High'].astype(float) - bitcoin_df_2['Low'].astype(float)
bitcoin_df_2['MA_5'] = bitcoin_df_2['Close'].rolling(window=5).mean()
bitcoin_df_2['MA_10'] = bitcoin_df_2['Close'].rolling(window=10).mean()
bitcoin_df_2['EMA_10'] = bitcoin_df_2['Close'].ewm(span=10, adjust=False).mean()
bitcoin_df_2['Volatility_5'] = bitcoin_df_2['Return'].rolling(window=5).std()
bitcoin_df_2['High'] = pd.to_numeric(bitcoin_df_2['High'], errors='coerce')
bitcoin_df_2['Low'] = pd.to_numeric(bitcoin_df_2['Low'], errors='coerce')
bitcoin_df_2['Volume'] = pd.to_numeric(bitcoin_df_2['Volume'], errors='coerce')
X = bitcoin_df_2[['Open', 'Volume', 'Spread', 'MA_5', 'MA_10', 'EMA_10', "Volatility_5"]]
X = X.fillna(0)
Y = bitcoin_df_2['Close']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
dates = bitcoin_df_2.loc[Y_test.index, 'Date']
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
reg = MLPRegressor(solver='lbfgs',alpha=1e-5,hidden_layer_sizes=(25, 10),random_state=1)
reg.fit(X_train, Y_train)
predictions = reg.predict(X_test)
combined_df = pd.DataFrame({'Date': dates, 'Actual': Y_test.values, 'Prediction': predictions})
combined_df.set_index('Date').sort_index()
combined_df['Year'] = combined_df['Date'].dt.year
sample = combined_df.head(200)
sample = sample.sort_index()
plt.plot(sample['Actual'], label = 'Actual',color = 'blue', linewidth = 1, alpha = 0.7)
plt.plot(sample['Prediction'], label = 'Predicted',color = 'Red', linewidth = 1, alpha = 0.7)
plt.title('Actual Price and Predicted Price')
plt.ylabel("Price at given point")
plt.xlabel('Observation')
plt.legend()
plt.show()
```
